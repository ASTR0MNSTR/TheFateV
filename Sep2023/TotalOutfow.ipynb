```python
import numpy as np
import matplotlib.pyplot as plt
import math
from __plt__ import *

def plotting_pars():
    plt.rcParams['font.size'] = 20
    mpl.rcParams['xtick.labelsize'] = 20
    mpl.rcParams['ytick.labelsize'] = 20 
    mpl.rcParams['axes.labelsize'] = 20
    mpl.rcParams['ytick.major.size'] = 15
    mpl.rcParams['xtick.major.size'] = 15
```

```python
ages = np.arange(9.01, 11.00, 0.01)
```

```python
a = 77.4
a_err = 8.52
# a_err = 20
b = 2.26*(10**9)
b_err = 0.23*(10**9)
T_0 = 1*(10**9)

av_out_on = np.log10(a*b*(math.exp(-(T_0)/b) - np.exp(-(10**ages)/b)))
av_out_on_up = np.log10((a+a_err)*(b+b_err)*(math.exp(-(T_0)/(b-b_err)) - np.exp(-(10**ages)/(b-b_err))))
av_out_on_down = np.log10((a-a_err)*(b-b_err)*(math.exp(-(T_0)/(b+b_err)) - np.exp(-(10**ages)/(b+b_err))))

a = 9.30
a_err = 0.81
b = 1.22*(10**9)
b_err = 0.04*(10**9)

av_out_off = np.log10(a*b*(math.exp(-(T_0)/b) - np.exp(-(10**ages)/b)))
av_out_off_up = np.log10((a+a_err)*(b+b_err)*(math.exp(-(T_0)/(b-b_err)) - np.exp(-(10**ages)/(b-b_err))))
av_out_off_down = np.log10((a-a_err)*(b-b_err)*(math.exp(-(T_0)/(b+b_err)) - np.exp(-(10**ages)/(b+b_err))))
```

```python
fig = plt.figure(figsize=(12, 6), tight_layout=True)
plotting_pars()
ax = fig.add_subplot()
ax.plot(ages, av_out_on, color = 'k', linestyle = ':', label = 'AGN ON')
ax.fill_between(ages, av_out_on_down, av_out_on_up, alpha = 0.2, color='grey')
ax.plot(ages, av_out_off, color = 'k', linestyle = '--', label = 'AGN OFF')
ax.fill_between(ages, av_out_off_down, av_out_off_up, alpha = 0.2, color='grey')

ax.tick_params(top=True, labeltop=False, bottom=True, labelbottom=True, right=True, direction='in')
ax.set_xlabel(r'$\log \mathrm{(age \: / \: yr)}$')
ax.set_ylabel(r'$\log \mathrm{(M_{tot. exp. H_2}/ \: M_\odot)}$')
ax.axhline(y = 9.954, color='k', linestyle = '-')
# ax.set_title('Start age = 1 Gyr')
ax.set_xlim(9.0, 10.1)
ax.set_xticks(np.arange(9.0, 10.2, 0.1))
ax.legend(fontsize=20)

#add plt.fill_between() function with a_err and b_err estimation

plt.savefig('./FIGURES/AV_OUT.pdf', dpi=300, transparent = True, bbox_inches = 'tight', pad_inches = 0.0001)
```
