```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

from __plt__ import *
```

```python
path_to_data = r'E:/databases/CIGALE/results.txt'
path_to_classes = r'E:/backup/backup_BPT/GAMA_ETG_OLA.csv'

DataFrameCigale = pd.read_csv(path_to_data, sep=r"\s+", header=0, engine='python')

usecols = ['SPECID', 'CATAID_1', 'BPT', 'WHAN']
DataFrameBPT = pd.read_csv(path_to_classes, usecols=usecols)
```

```python
DataFrameAGN005 = DataFrameCigale[DataFrameCigale['bayes.agn.fracAGN'] - 2*DataFrameCigale['bayes.agn.fracAGN_err'] > 0.05]

DataFrame005 = pd.merge(DataFrameBPT, DataFrameAGN005, how='inner', on='CATAID_1')
DataFrame005['CIGALE_AGN'] = True
print(DataFrame005.shape)
DataFrame005.to_csv('CIGALE_AGN_005.csv', index=False)
```

```python
gs_top = plt.GridSpec(1, 2, hspace=0, wspace=0)
fig = plt.figure(figsize=(12, 6), tight_layout=True)

plt.rcParams['font.size'] = 15

ax1 = fig.add_subplot(gs_top[0,0])
ax2 = fig.add_subplot(gs_top[0,1])
fig.suptitle('bayes.agn.fracAGN - 2*fracAGN_err > 0.05, 50 gal.')

ax1 = plotting_round_histo_BPT(ax1, DataFrame005['BPT'])

ax2 = plotting_round_histo_WHAN(ax2, DataFrame005['WHAN'])

fig.savefig('DIAG_CIGALE_AGN_005.pdf', dpi=300, transparent = True, bbox_inches = 'tight', pad_inches = 0.0001)
```

```python
DataFrameAGN005 = DataFrameCigale[DataFrameCigale['bayes.agn.fracAGN'] - 2*DataFrameCigale['bayes.agn.fracAGN_err'] > 0.01]

DataFrame001 = pd.merge(DataFrameBPT, DataFrameAGN005, how='inner', on='CATAID_1')

print(DataFrame001.shape)
DataFrame001.to_csv('CIGALE_AGN_001.csv', index=False)
```
