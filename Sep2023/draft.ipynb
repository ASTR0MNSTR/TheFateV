```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
ages = np.arange(9.01, 11.25, 0.01)
```

```python
a = 77.31
a_err = 8.6
# a_err = 20
b = 2.26*(10**9)
b_err = 0.23*(10**9)
T_0 = 1*(10**9)

av_out_on = np.log10(a*b*(math.exp(-(T_0)/b) - np.exp(-(10**ages)/b))/(10**ages - T_0))
av_out_on_up = np.log10((a+a_err)*(b+b_err)*(math.exp(-(T_0)/(b-b_err)) - np.exp(-(10**ages)/(b-b_err)))/(10**ages - T_0))
av_out_on_down = np.log10((a-a_err)*(b-b_err)*(math.exp(-(T_0)/(b+b_err)) - np.exp(-(10**ages)/(b+b_err)))/(10**ages - T_0))

a = 9.27
a_err = 0.79
b = 1.22*(10**9)
b_err = 0.04*(10**9)

av_out_off = np.log10(a*b*(math.exp(-(T_0)/b) - np.exp(-(10**ages)/b))/(10**ages - T_0))
av_out_off_up = np.log10((a+a_err)*(b+b_err)*(math.exp(-(T_0)/(b-b_err)) - np.exp(-(10**ages)/(b-b_err)))/(10**ages - T_0))
av_out_off_down = np.log10((a-a_err)*(b-b_err)*(math.exp(-(T_0)/(b+b_err)) - np.exp(-(10**ages)/(b+b_err)))/(10**ages - T_0))
```

```python
fig, ax = plt.subplots(1, 1)

ax.plot(ages, av_out_on, color = 'k', linestyle = ':', label = 'AGN ON')
ax.fill_between(ages, av_out_on_down, av_out_on_up, alpha = 0.2)
ax.plot(ages, av_out_off, color = 'k', linestyle = '--', label = 'AGN OFF')
ax.fill_between(ages, av_out_off_down, av_out_off_up, alpha = 0.2)

ax.set_xlabel(r'$log(age/yr)$')
ax.set_ylabel(r'$log(outflow rate/ \: M_\odot \: yr^{-1})$')
ax.axhline(y = 0, color='k', linestyle = '-')
ax.set_title('Start age = 1 Gyr')
ax.set_xlim(9, 11.25)
ax.legend()

#add plt.fill_between() function with a_err and b_err estimation

plt.savefig('./FIGURES/AV_OUT.pdf')
```
